```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_2i = pd.read_csv('WT_26_KO_6_2i_overlap_Repeatmasker.interval', sep='\t', header=None, names=['Chr', 'Start', 'End', 'WT_26_2i', 'KO_6_2i', 'TE'], usecols=[0,1,2,3,9,15])
df_serum = pd.read_csv('WT_26_KO_6_serum_overlap_Repeatmasker.interval', sep='\t', header=None, names=['Chr', 'Start', 'End', 'WT_26_serum', 'KO_6_serum', 'TE'], usecols=[0,1,2,3,9,15])
```

```python
TE_count = df_2i.TE.value_counts()
print(TE_count.head(100))     
```

```python
name_list = ['WT(serum)', 'KO(serum)', 'WT(2i)', 'KO(2i)']
TE_list = ['MMETn-int', 'RLTR4_MM-int', 'MMERGLN-int', 'IAPEz-int', 'L1Md_T']

for TE in TE_list:
    df_filtered_serum = df_serum[df_serum['TE'] == TE]
    df_filtered_2i = df_2i[df_2i['TE'] == TE]

    plt.boxplot([df_filtered_serum['WT_26_serum'], df_filtered_serum['KO_6_serum'], df_filtered_2i['WT_26_2i'], df_filtered_2i['KO_6_2i']], labels=name_list)
    plt.title(TE)
    plt.show()
    plt.close()
```
